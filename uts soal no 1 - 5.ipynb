```python
company_name_list = [{'name': 'Company 1'},
          {'name': 'Company 2'},
          {'name': 'Company 3'}]

employee_name_list = [{'name': 'John Doe'},
          {'name': 'Tom Smith'},
          {'name': 'Andrew Sebastian'}]

company_detail_list = {
      'Company 1': {
          'name': 'Company 1',
          'domain': 'Retail',
          'clients': [
              {
                  'name': 'acme.inc',
                  'country': 'united states'
              },
              {
                  'name': 'Wayne.co',
                  'country': 'united states'
              }
          ]
      },
      'Company 2': {
          'name': 'Company 2',
          'domain': 'Construction',
          'clients': [
              {
                  'name': 'Tesla',
                  'country': 'united states'
              },
              {
                  'name': 'Japan Airlines',
                  'country': 'japan'
              },
              {
                  'name': 'Indofood',
                  'country': 'indonesia'
              }
          ]
      },
      'Company 3': {
          'name': 'Company 3',
          'domain': 'Healthcare',
          'clients': [
              {
                  'name': 'Petronas',
                  'country': 'malaysia'
              },
              {
                  'name': 'VW Group',
                  'country': 'germany'
              },
              {
                  'name': 'IBM',
                  'country': 'united states'
              },
              {
                  'name': 'Mitsubishi',
                  'country': 'japan'
              }
          ]
      }
  }

employee_detail_list = {
      'John Doe': {
          'name': 'EMP-0001',
          'first_name': 'John',
          'last_name': 'Doe',
          'full_name': 'John Doe',
          'company': 'Company 1'
      },
      'Tom Smith': {
          'name': 'EMP-0002',
          'first_name': 'Tom',
          'last_name': 'Smith',
          'full_name': 'Tom Smith',
          'company': 'Company 2'
      },
      'Andrew Sebastian': {
          'name': 'EMP-0003',
          'first_name': 'Andrew',
          'last_name': 'Sebastian',
          'full_name': 'Andrew Sebastian',
          'company': 'Company 2'
      },
  }

def sort_company():
    company_names = [company['name'] for company in company_name_list]

    sorted_company_names = sorted(company_names)

    print("Daftar Perusahaan yang Telah Diurutkan:")
    for company_name in sorted_company_names:
        print(company_name)

sort_company();
```

```python
def get_company_domain(company_name):
    company_detail = company_detail_list.get(company_name)
    if company_detail:
        return company_detail['domain']
    else:
        return "Perusahaan tidak ditemukan."

company_name = "Company 3"
company_domain = get_company_domain(company_name)
print("Domain dari", company_name, "adalah:", company_domain)
```

```python
def get_employees():
    result = []

    for employee_name, employee_detail in employee_detail_list.items():
        full_name = employee_detail['full_name']
        company_name = employee_detail['company']

        company_domain = company_detail_list[company_name]['domain']

        result.append({
            "full_name": full_name,
            "company": company_name,
            "domain": company_domain
        })

    return result

get_employees()
```

```python
def get_employees_by_company():
    result = []

    for company_info in company_name_list:
        company_name = company_info['name']
        employees = []

        for employee_name, employee_detail in employee_detail_list.items():
            if employee_detail['company'] == company_name:
                employees.append(employee_name)

        result.append({
            "company": company_name,
            "employees": employees
        })

    return result

get_employees_by_company()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler

data_startup = pd.read_csv('50_Startups.csv')

startup_table = data_startup[['R&D Spend', 'Administration', 'Marketing Spend', 'State', 'Profit']]

print(startup_table.sample(7))

data_hilang = startup_table.isnull().sum()
print("Data yang kosong (NaN):")
print(data_hilang)

startup_table[['R&D Spend', 'Administration', 'Marketing Spend', 'Profit']] = startup_table[['R&D Spend', 'Administration', 'Marketing Spend', 'Profit']].fillna(startup_table[['R&D Spend', 'Administration', 'Marketing Spend', 'Profit']].mean())

label_encoder = LabelEncoder()
startup_table['State'] = label_encoder.fit_transform(startup_table['State'])

onehot_encoder = OneHotEncoder()
encode = onehot_encoder.fit_transform(startup_table[['State']])
endcode_df = pd.DataFrame(encode.toarray(), columns=label_encoder.classes_)
startup_table = pd.concat([startup_table, endcode_df], axis=1)
startup_table.drop('State', axis=1, inplace=True)

startup_table['Tax'] = (startup_table['Profit'] + startup_table['Marketing Spend'] + startup_table['Administration']) * 0.05

scaler = StandardScaler()
kolom = ['R&D Spend', 'Administration', 'Marketing Spend', 'Profit', 'Tax']
startup_table[kolom] = scaler.fit_transform(startup_table[kolom])

print("Data setelah pre-processing:")
print(startup_table.head())
```
